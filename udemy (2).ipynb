```python
#optimization with numpy

import numpy as np
array = np.random.rand(1000000)
def sum_with_loop(array):
  total = 0
  for num in array:
    total += num
  return total
def sum_with_numpy(array):
  return np.sum(array)
print("Sum with loop:",sum_with_loop(array))
print("Sum with numpy:",sum_with_numpy(array))
```
